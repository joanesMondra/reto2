```python
import os
import shutil

# Rutas de las carpetas
ruta_origen = r"C:\Users\joane\PycharmProjects\Master_ia\reto2\Imagenes_defectos"
ruta_destino = r"C:\Users\joane\PycharmProjects\Master_ia\reto2\imagenes_label"

# Crear la carpeta de destino si no existe
os.makedirs(ruta_destino, exist_ok=True)

# Contador global para garantizar unicidad
contador = 0

# Recorrer todas las subcarpetas de la carpeta origen
for subcarpeta, _, archivos in os.walk(ruta_origen):
    subcarpeta_nombre = os.path.basename(subcarpeta)  # Nombre de la subcarpeta
    for archivo in archivos:
        if archivo.lower().endswith(".bmp"):  # Filtrar archivos BMP (insensible a mayúsculas)
            ruta_archivo = os.path.join(subcarpeta, archivo)

            # Crear un nuevo nombre único para el archivo
            nuevo_nombre = f"{subcarpeta_nombre}_{contador}_{archivo}"
            ruta_nueva = os.path.join(ruta_destino, nuevo_nombre)

            # Mover el archivo y actualizar el contador
            shutil.move(ruta_archivo, ruta_nueva)
            print(f"Movido: {ruta_archivo} -> {ruta_nueva}")
            contador += 1

print(f"Se han movido {contador} imágenes .bmp a {ruta_destino}")
```
