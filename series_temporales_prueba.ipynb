```python
import pandas as pd

# Cargar el archivo CSV para inspeccionar su contenido
file_path = './bearing_fault_detection_reduced/normal/12.288.csv'
data = pd.read_csv(file_path)

# Mostrar las primeras filas del archivo para comprender su estructura
data.head(), data.columns
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Renombrar columnas basándose en la suposición de 8 señales distintas
data.columns = [
    "tachometer",
    "acc_under_axial",
    "acc_under_radiale",
    "acc_under_tangencial",
    "acc_over_axial",
    "acc_over_radiale",
    "acc_over_tangencial",
    "microphone",
]

# Crear los gráficos de cada señal
fig, axes = plt.subplots(2, 4, figsize=(15, 8))
axes = axes.ravel()

for i, column in enumerate(data.columns):
    axes[i].plot(data[column], label=column, linewidth=0.8)
    axes[i].set_title(column)
    axes[i].set_xlabel("x-label")
    axes[i].set_ylabel("y-label")
    axes[i].legend()

plt.tight_layout()
plt.show()
```
